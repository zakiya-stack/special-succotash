```python
from sklearn.decomposition import PCA
```

```python
def pca_analysis(n_components, data):
    pca = PCA(n_components)
    X_pca = pca.fit_transform(data)
    return pca, X_pca
```

```python
azdias_pca, azdias_X_pca = pca_analysis(None, azdias_scaled)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
def k_means_trial(data, center):
    kmeans = KMeans(n_clusters = center, random_state = 23, n_jobs = -1)
    model = kmeans.fit(data)
    score = np.abs(model.score(data))
    return score
```

```python
import numpy as np
list_score =  []
centers = list(range(1,10))
for center in centers:
    list_score.append(k_means_trial(azdias_X_pca_after_choosing_200, center))
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(7, 6)) 
plt.plot(centers, list_score, color='green', marker='o', linestyle='dashed');
matplotlib.rc('xtick', labelsize=20)  
matplotlib.rc('ytick', labelsize=20)
plt.xlabel('K', fontsize=10)
plt.ylabel('SSE', fontsize=10)
plt.title('K', fontsize=10)
plt.show()
```
